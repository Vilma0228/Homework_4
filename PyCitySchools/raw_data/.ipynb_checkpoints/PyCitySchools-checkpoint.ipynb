```python
import pandas as pd
import os
import numpy as np
import csv
```

```python
# The path to our CSV file
school_file = "schools_complete.csv"
students_file = "students_complete.csv"

# Read our data into pandas
school_df = pd.read_csv(school_file)
students_df = pd.read_csv(students_file)
```

```python
#District Summary = School_df & Students_df

# Create a high level snapshot (in table form) of the district's key metrics, including:
# Total Schools
Total_Schools = 0
Total_Schools = school_df['school'].count()

# #   Total Students
Total_Students = 0
Total_Students = school_df['size'].sum()

# # #   Total Budget
Total_Budget = 0
Total_Budget = school_df['budget'].sum()

# # #   Average Math Score 
Average_Math_Score = 0
Average_Math_Score  = students_df['math_score'].mean()

# # #   Average Reading Score
Average_Reading_Score = 0
Average_Reading_Score = students_df['reading_score'].mean()
```

```python
#Part 1 = % Passing Math (% of Score >=60) 
# ds_math_pass_count = 0
# ds_math_pass_count = students_df[students_df['math_score'] >59.9].count()
# ds_math_pass_count

# Percent_Passing_Math = 0
# Percent_Passing_Math = (ds_math_pass_count / Total_Students) *100

students_df['check_math_score'] = students_df['math_score'].apply(lambda x: x > 59)
students_df['check_math_score'].value_counts()
```

```python
not_passing_math = students_df['check_math_score'].value_counts()[0]
passing_math = students_df['check_math_score'].value_counts()[1]
total_tests = passing_math + not_passing_math

percent_math = passing_math / total_tests
percent_math
```

```python
# students_df['check_reading_score'] 
students_df['check_reading_score'] = students_df['reading_score'].apply(lambda x: x > 59)
```

```python
students_df.head()
```

```python
passing_reading = len(students_df[students_df['check_reading_score'] == True])

passing_reading
```

```python
percent_reading = passing_reading / total_tests
percent_reading
```

```python
#  * Overall Passing Rate 

students_df.apply(lambda row: (row.reading_score + row.math_score)/2, axis =1)

students_df['Average Score'] = students_df.apply(lambda row: (row.reading_score + row.math_score)/2, axis=1)

students_df.head()
```

```python
Total_Average_Score = students_df['Average Score'] .sum()
Overall_Passing_Rate = Total_Average_Score / total_tests
Overall_Passing_Rate
```

```python
District_Summary = pd.DataFrame(
{
    'Total_Schools':[Total_Schools], 
    'Total_Students': [Total_Students],
    'Total_Budget': [Total_Budget],
    'Average_Math_Score': [Average_Math_Score],
    'Average_Reading_Score': [Average_Reading_Score],
    '% Passing Math': [percent_math],
    '% Passing Reading': [percent_reading],
    'Overall Passing Rate': [Overall_Passing_Rate]
})

District_Summary
```
