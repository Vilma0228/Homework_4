```python
#District Summary

# Create a high level snapshot (in table form) of the district's key metrics, including:

ds_school = school_df['name'].count()

ds_students= school_df['size'].sum()

ds_budget = school_df['budget'].sum()

ds_math = students_df['math_score'].mean()

ds_reading = students_df['reading_score'].mean()

ds_math_pass = students_df['math_score']
```

```python
#Determine range for Pass or Fail

math_df = pd.DataFrame(students_df, columns = ['name', 'math_score'])

math = []

for row in math_df['math_score']: 
    if row > 75:
        math.append('Pass')
        
    else:
        math.append('Fail')
        
math_df['math'] = math

math_df.head()
```

```python
math_total = math_df["name"].nunique()
math_total
```

```python
math_df["math"] == str.count("Pass") #"Pass".count()
```

```python
#Group by Pass or Fail
math_group = math_df.groupby("math")

# Count how many Pass or Fail scores and create DataFrame of the data

math_result = pd.DataFrame(math_group["math"].value_counts())
math_result
```
